```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
search_with_feature = 120000
search_without_feature = 130000
total_sessions_with = 300000
total_sessions_without = 200000
cart_with_feature = 5406
cart_without_feature = 5117
```

```python
search = [search_without_feature, search_with_feature]
sessions = [total_sessions_without, total_sessions_with]
```

```python
z_search, pval_search = proportions_ztest(search, sessions)

z_search, pval_search
```

```python
cart = [cart_without_feature, cart_with_feature]
z_cart, pval_cart = proportions_ztest(cart, search)

z_cart, pval_cart
```
